# Get embeddings for words in the SAC

```python
from ws_client import WebSocketClient 
```

```python
emb = WebSocketClient("ws://localhost:8765/")
```

```python
emb["تجربة"]
```

```python
w = open("words")
```

```python
ww = w.readlines()
```

```python
len(ww)
```

```python
embs = [emb[www.strip("\n")] for www in ww]
```

```python
out = open("words.vec", "w")
```

```python
out.write(str(len(embs))+" "+str(emb.vector_size)+"\n")
```

```python
for w,e in zip(ww,embs):
    out.write(w.strip("\n")+" "+ " ".join([str(ee) for ee in e])+"\n")
```

```python
out.flush()
```

```python
out.close()
```
